
**Introducción**

En este documento se presenta un modelo 1-D de evaporación de líquido criogénico en un regimen isobárico. Se utilizaron las ecuaciones del modelo de  Huerta (2019), modificando la geometría de las ecuaciones para modelar el comportamiento en un estanque cilíndrico horizontal.
A continuación, se muestran las ecuaciones del modelo.



**1. Geometría del modelo**

En primer lugar, se presentan las ecuaciones que modelan la física del problema. A continuación, se muestra una imagen del estanque, visto desde el lado.
Como se puede ver, existe una fase líquida y una fase vapor. Para cada una de las fases, se modela un ingreso de calor por separado. Además existe una fuente de calor extra proveniente del piso del estanque. 
En el almacenamiento, el calor provoca que cierta parte del líquido se evapore. Para mantener presiones seguras dentro del estanque, un flujo de vapor se libera en la parte superior del estanque como Boil Off Gas (BOG).


![Vista desde el lado](vista_frontal.png)

El estanque modelado cuenta con tapas planas. Es decir, en las tapas del estanque se cuenta con circunferencias. La vista desde esa perspectiva se puede ver a continuación. 

![Vista tapas](vista_tapas.png)

Como se puede apreciar en la figura, la circunferencia cuenta con fase líquida y fase vapor. Es importante destacar el ángulo alpha, que se forma por el diámetro de la circunferencia en 90°, y la altura del líquido. A medida que aumenta la altura del líquido, este ángulo será mayor. 

A continuación, se muestran las ecuaciones que representan el modelo geométrico de ambas fases:

**Area del liquido y del vapor en las tapas circulares (1)**

La vista lateral del cilindro corresponde a una circunferencia, en la cual dependiendo del nivel de llenado de líquido (su altura h desde la base) se puede definir el área tanto del líquido como del gas:

$$
A_{L}^{\text{tapa}}(h) \;=\; \,\big(R^2arcos(\frac{R-h}{R})- (R-h)\cdot \sqrt{2Rh-h^2} ),\\
A_{V}^{\text{tapa}}(h) = \pi R^{2}-A_{L}^{\text{transv}}(h)
$$

**Volumen del líquido (2)**

El volúmen del liquido corresponde al área del líquido 
visto desde la tapa (1) multipicada por el largo del cilindro L.

$$

V_{L}^{\text{cilindro}}(h) \;=\; L\,\big(R^2arcos(\frac{R-h}{R})- (R-h)\cdot \sqrt{2Rh-h^2} )

$$

Por otro lado, el volumen del vapor será el volumen total menos el del líquido:

$$
V_{V}=L\big(\pi R^{2}-A_{L}^{\text{transv}}(h)\big) 
$$

**Perimetros de liquido y vapor en las tapas (3)**

El perímetro mojado en liquido y vapor en las tapas se definen mediante las siguientes expresiones:

$$
s_{L}(h)=2R\,\theta(h),\quad
s_{V}(h)=2\pi R - 2R\,\theta(h)
$$

Donde al ángulo central de la circunferencia corresponde a:

$$
\theta(h) \;=\; \,\arccos\!\left(\frac{R-h}{R}\right)
$$

**Áreas de pared mojadas por cada fase (4)**

Corresponden al área de la pared del estanque que están en contacto con el líquido y el vapor. Se calculan como el perímetro mojado en las tapas, multiplicado por el largo del estanque, más el área correspondiente a las tapas.

$$
A_{w,L}=L\,s_{L}(h)+2\cdot A_{L}^{\text{transv}}(h) ,\qquad
A_{w,V}=L\,s_{V}(h)+ 2\cdot A_{V}^{\text{transv}}(h)
$$

**Largo de cuerda en la interfase LV (5)**

Corresponde al largo de la superficie de la interfase, si se mira desde el lado de las tapas. Notar que en h = R, este largo es máximo y corrresponde al diametro de la tapa. En h = 0 y h = R este largo es 0 ya que no hay interfase.

$$
c(h)\;=\; \; 2\sqrt{\,2Rh-h^2\,},\\
$$

**Área interfacial líquido–vapor (6)**

Corresponde al área total de la interfase:

$$
A_{i}(h)=2L\sqrt{\,2Rh-h^2 \,}
$$

**Razones área/volumen para fuentes volumétricas (7)**

$$
a_{L}(h)=\frac{A_{w,L}}{V_{L}},\qquad
a_{V}(h)=\frac{A_{w,V}}{V_{V}}.
$$

**2. Balances de masa y energía**

A continuación, se muestra el modelo termodinámico. Se muestran los distintos balances de masa y energía que componen el modelo.

**Balance de masa del líquido (8)**

Siendo $B_L$ la masa evaporada de liquido:

$$
\frac{\mathrm{d}}{\mathrm{d}t}\big(\rho_{L}V_{L}\big)=-\dot B_{L}.
$$

**Balance global de masa y extracción de BOG (9)**

En operación isobárica el gas producido no se acumula libremente; para mantener la presión constante se extrae un caudal $\dot B$ de la parte superior (BOG). El balance total del tanque (líquido+vapor) relaciona la evaporación con la acumulación en el espacio de vapor y la extracción de BOG.

$$
-\dot B=\frac{\mathrm{d}}{\mathrm{d}t}\big(\rho_{L}V_{L}+\rho_{V}V_{V}\big).
$$

**Balance de energía del líquido (10)**

El líquido recibe calor desde la pared sumergida (lateral y tapas), desde la losa/apoyos y desde el vapor supercalentado en la interfase. Ese calor se reparte entre (i) acumular entalpía sensible del líquido y (ii) *proveer calor de vaporización* a la masa que pasa a vapor ($\dot B_L h_V$).

$$
U_{s,L}\,A_{w,L}\,[T_{\infty}-T_{L}]
+U_{e,L}\,2A_{L}^{\text{transv}}(h)\,[T_{\infty}-T_{L}]
+\dot Q_{\text{slab}}(h)\\
\qquad+\dot Q_{VL}-\dot B_{L}\,h_{V}(T_{L})
=\frac{\mathrm{d}}{\mathrm{d}t}\big[\rho_{L}V_{L}\,h_{L}(T_{L})\big].
$$

**3. Modelo de transferencia de calor**

**Flujo interfacial vapor a líquido (conducción) (11)**

El vapor, más caliente que el líquido en no equilibrio térmico, transfiere calor por conducción a través de la interfase plana. Se evalúa con la ley de Fourier en $z=h$ (interfase).

$$
\dot Q_{VL}(t)=-\,k_{V}\,A_{i}(h)\,
\left.\frac{\partial T_{V}}{\partial z}\right|_{z=h}.
$$

**Ecuación de energía para el vapor (vertical) (12)**

En el vapor dominan (i) difusión térmica vertical (conducción), (ii) advección vertical debida al flujo ascendente del gas evaporado, y (iii) un término fuente volumétrico que representa el calentamiento desde la pared (convección externa + conducción del aislamiento, promediado como $U_V$ y convertido a fuente por $a_V$, que es la razón área-volumen).

$$
\frac{\partial T_V}{\partial t}
= \alpha_V\!\left[
\frac{\partial^2 T_V}{\partial z^2}
+ \frac{A'(z)}{A(z)}\,\frac{\partial T_V}{\partial z}
\right]
- v_z(z)\,\frac{\partial T_V}{\partial z}
+ \frac{\dot{s}_{\mathrm{wall}}(z,T_V)}{\rho_V\,c_{p,V}} \, .
$$

**Velocidad media de advección en el vapor (13)**

Se aproxima el campo de velocidades por una velocidad vertical media que surge del caudal evaporado por unidad de área interfacial y densidad de vapor (aproximación 1-D).

$$
v_{z}(h)=\frac{\dot B_{L}/\rho_{V}}{A_{V}^{\text{transv}}(h)}.
$$

**Condiciones de borde para la PDE del vapor (14)**

(i) continuidad de temperatura en la interfase ($T_V=T_L$), y (ii) techo adiabático (sin flujo de calor por la parte superior).

$$
T_{V}(t,0)=T_{L}(t),\qquad
\left.\frac{\partial T_{V}}{\partial z}\right|_{z=l_{V}^{\text{vert}}(t)}=0.
$$

**Calor por pared hacia el líquido (15)**

Hay aporte de calor desde la superficie lateral mojada y desde las tapas en contacto con el líquido (coeficientes globales $U_{s,L}$ y $U_{e,L}$).

$$
\dot Q_{L}
=U_{s,L}\,L\,s_{L}(h)\,[T_{\infty}-T_{L}]
+U_{e,L}\,2A_{L}^{\text{cap}}(h)\,[T_{\infty}-T_{L}].
$$

**Calor por pared hacia el vapor (16)**

Balance superficial: aporte desde la superficie lateral mojada y desde las tapas en contacto con el vapor (coeficientes globales $U_{s,V}$ y $U_{e,V}$).

$$
\dot Q_{V}
=U_{s,V}\,L\,s_{V}(h)\,[T_{\infty}-\bar T_{V}(t)]
+U_{e,V}\,2\big(\pi R^{2}-A_{L}(h)\big)\,[T_{\infty}-\bar T_{V}(t)].
$$

**Relación de la extracción total (17)**

La extracción neta de BOG se obtiene como la evaporación menos la tasa de acumulación de masa en el espacio de vapor (si el vapor se expande/contrae).

$$
\dot B=\dot B_{L}-\frac{\mathrm{d}}{\mathrm{d}t}\big(\rho_{V}V_{V}\big).
$$
